```python
from typing import List

def twoSum(nums: List[int], target: int) -> List[int]:
    for i in range(0, len(nums)):
        for j in range(0, len(nums)):
            if i == j:
                continue
            elif nums[i] + nums[j] == target:
                return [i, j]

print(twoSum([3, 4, 5, 6], 11))          
```

```python
from collections import defaultdict
from typing import List

def groupAnagrams(strs: List[str]) -> List[List[str]]:
    res = defaultdict(list)
    for i in strs:
        sortedS = ''.join(sorted(i))
        res[sortedS].append(i)
    return list(res.values())

print(groupAnagrams(["act","pots","tops","cat","stop","hat"]))
```

```python
def productExceptSelf(nums: List[int]) -> List[int]:
    new_arr = []
    for i in range(0, len(nums)):
        product = 1
        for j in range(0, len(nums)):
            if i != j:
                product *= nums[j]
        new_arr.append(product)
    return new_arr

print(productExceptSelf([1,2,4,6]))
```

```python
from collections import defaultdict

class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        rows_map = defaultdict(list)
        cols_map = defaultdict(list)
        square_map = defaultdict(list)

        for r in range(9):
            for c in range(9):
                if board[r][c] == ".":
                    continue
                if board[r][c] in rows_map[r] or board[r][c] in cols_map[c] or board[r][c] in square_map[(r//3, c//3)]:
                    return False
                else:
                    rows_map[r].append(board[r][c])
                    cols_map[c].append(board[r][c])
                    square_map[(r//3, c//3)].append(board[r][c])
        return True

board = [["1","2",".",".","3",".",".",".","."],
 ["4",".",".","5",".",".",".",".","."],
 [".","9","8",".",".",".",".",".","3"],
 ["5",".",".",".","6",".",".",".","4"],
 [".",".",".","8",".","3",".",".","5"],
 ["7",".",".",".","2",".",".",".","6"],
 [".",".",".",".",".",".","2",".","."],
 [".",".",".","4","1","9",".",".","8"],
 [".",".",".",".","8",".",".","7","9"]]

a = Solution()
print(a.isValidSudoku(board))
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        newStr = ''
        for c in s:
            if c.isalnum():
                newStr += c.lower()
        return newStr == newStr[::-1]
    
a = Solution()
s = "Was it a car or a cat I saw?"
print(a.isPalindrome(s))
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        final_ans = set()
        for i in range(len(nums)):
            for j in range(len(nums)):
                for k in range(len(nums)):
                    if i != j and j != k and i != k:
                        triplet = tuple(sorted([nums[i],nums[j],nums[k]]))
                        if sum(triplet) == 0:
                            final_ans.add(triplet)
        return [list(a) for a in final_ans]

nums = [-1,0,1,2,-1,-4]
a = Solution()
print(a.threeSum(nums))
```

```python
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        print(heights)
        max_area = 0
        l, r = 0, len(heights)-1
        while l < r:
            while l < r:
                cur_area = min(heights[l], heights[r]) * (r-l)
                max_area = max(cur_area, max_area)
                r-=1
            l+=1
            r = len(heights)-1
        return max_area
a = Solution()
print(a.maxArea(heights=[1,7,2,5,4,7,3,6]))
```

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        #formula - min(l,r) + i
        n = len(height)
        max_left = [0] * n
        max_right = [0] * n
        traped_water = [0] * n
        for i, j in enumerate(height):
            if i == 0:
                max_left[i] = j
                max_right[i] = max(height[i+1:])
                traped_water[i] = min(max_left[i], max_right[i]) - j
            elif i == len(height) - 1:
                max_left[i] = max(height[:i])
                max_right[i] = height[i]
                traped_water[i] = min(max_left[i], max_right[i]) - j
            else:
                max_left[i] = max(height[:i])
                max_right[i] = max(height[i+1:])
                traped_water[i] = min(max_left[i], max_right[i]) - j
        
        traped_water = [x if x >= 0 else 0 for x in traped_water]
        return sum(traped_water)


a = Solution()
print(a.trap(height= [0,2,0,3,1,0,1,3,2,1]))
```

```python
class MinStack:

    def __init__(self):
        arr = []

    def push(self, val: int) -> None:
        arr.append(val)

    def pop(self) -> None:
        arr.pop()

    def top(self) -> int:
        return arr[-1]

    def getMin(self) -> int:
        return sorted(arr)[0]

minStack = MinStack()
minStack.push(1)
minStack.push(2)
minStack.push(0)
print(minStack.getMin())
minStack.pop()
print(minStack.top())
print(minStack.getMin())
```

```python
from typing import List

class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = []
        for j in tokens:
            if j not in "+-*/":
                stack.append(j)
            else:
                b = int(stack.pop())
                a = int(stack.pop())
                if j == "+":
                    stack.append(a+b)
                if j == "-":
                    stack.append(a-b)
                if j == "*":
                    stack.append(a*b)
                if j == "/":
                    stack.append(a/b)
        return int(stack[0])
a = Solution()
print(a.evalRPN(tokens = ["1","2","+","3","*","4","-"]))
```

```python
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        max_temp = 0
        result = [0]*len(temperatures)
        for i, j in enumerate(temperatures):
            for temps in range(i+1, len(temperatures)):
                if temperatures[temps] > j:
                    result[i] = temps - i
                    break
        return result

a = Solution()
print(a.dailyTemperatures(temperatures = [30,38,30,36,35,40,28]))
#Output: [1,4,1,2,1,0,0]
```
